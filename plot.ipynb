```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def is_contains_chinese(strs):
    for _char in strs:
        if '\u4e00' <= _char <= '\u9fa5':
            return True
    return False

import matplotlib
matplotlib.matplotlib_fname()
```

```python
f = open('log_abs.txt', 'r')
meta = f.readline().strip().split()
print("Date:\t", meta[0].strip('#'))
print("Time:\t", meta[1])
print("VIN:\t", meta[2])

enabled = ['可用','已操作','激活']
disabled = ['不可用','未操作','未激活']
errors = ['ERROR', '-']

# attrs = f.readline().strip().split()
attrs = f.readline()
attrs = [x.strip('#') for x in f.readline().strip().split()]
print(attrs)
d = {}
for i in attrs:
    d[i] = []
units = [x.strip('#') for x in f.readline().strip().split()]
print(units)
line = f.readline()
while line != '':
    content = line.strip().split()
    if 'ERROR' not in line and '故障' not in line and len(content) == len(attrs):
        for i in range(len(attrs)):
            if is_contains_chinese(content[i]):
                if(content[i] in enabled):
                    d[attrs[i]].append(1)
                elif (content[i] in disabled):
                    d[attrs[i]].append(0)
                else:
                    d[attrs[i]].append(-1)
            elif content[i] in errors:
                d[attrs[i]].append(-1)
            else:
                d[attrs[i]].append(float(content[i]))
    line = f.readline()
for i in attrs:
    print(i, len(d[i]))
```

```python

for i in attrs:
    if(i != 'Time'):
        plt.figure(figsize=(25,5))
        plt.plot(d['Time'], d[i])
        plt.legend([i])
        plt.show()
#plt.legend(attrs[1:])
```
